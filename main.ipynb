```python
# Importing Modules
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading Data
train_data = pd.read_csv("data.csv", error_bad_lines=False)
train_data.head()
```

```python
# Type of Passwords
train_data["strength"].uniqueque()
```

```python
# Finding null values
train_data.isna().sum()
```

```python
# Removing null values
train_data.dropna(inplace=True)
train_data.isna().sum()
```

```python
# Plotting Data
sns.countplot(train_data["strength"])
```

```python
# Making an array of passwords
passwords_tuple = np.array(train_data)
passwords_tuple
```

```python
# Randomly Shuffling Passwords for Robustness
import random
random.shuffle(passwords_tuple)
```

```python
# Seperating Columns
x = [item[0] for item in passwords_tuple]
y = [item[1] for item in passwords_tuple]
```

```python
x
```

```python
y
```

```python
# Using TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=lambda x: [i for i in x])
```

```python
# Applying Vectorizer
X = vectorizer.fit_transform(x)
X.shape
```

```python
# Printing all features
vectorizer.get_feature_names()
```

```python
# Storing into dataframe
first_document_vector = X[0]
df = pd.DataFrame(first_document_vector.T.todense(), index=vectorizer.get_feature_names(), columns=["TF-IDF"])
df.sort_values(by=["TF-IDF"], ascending=False, inplace=True)
df
```

```python
# Splitting into train & test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
